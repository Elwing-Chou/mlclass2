```python
def nameflow(name):
    n = name.split(",")[1].split(".")[0].replace(" ", "")
    if n == "Mr":
        return "Mr"
    elif n == "Mrs":
        return "Mrs"
    elif n == "Miss":
        return "Miss"
    else:
        return "Unknown"
train_df["Name"] = train_df["Name"].apply(nameflow)
d = pd.get_dummies(train_df["Name"])
train_df = pd.concat([train_df, d], axis=1)
train_df
```

```python
train_df = train_df.drop(["PassengerId", "Name", "Sex", "Ticket" , "Cabin", "Embarked", "Unknown"], axis=1)
```

```python
train_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14, 10))
sns.heatmap(train_df.corr(), annot=True, cmap="BrBG")
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=5)
clf.fit(train_df.drop(["Survived"], axis=1), train_df["Survived"])
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(clf, out_file=None, feature_names=train_df.drop(["Survived"], axis=1).columns,
                    class_names=["Not Survived", "Survived"], filled=True, special_characters=True)
graph = graphviz.Source(g)
graph
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=23, max_depth=7)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
scores = cross_val_score(clf, train_df.drop(["Survived"], axis=1), 
                         train_df["Survived"], cv=10)
print("åæ¬¡:", scores)
print("å¹³å‡:", np.average(scores))
```

```python
test_df = test_df.fillna(med)
test_df["Embarked"] = test_df["Embarked"].fillna(most)
d = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, d], axis=1)
d = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, d], axis=1)
test_df["Name"] = test_df["Name"].apply(nameflow)
d = pd.get_dummies(test_df["Name"])
test_df = pd.concat([test_df, d], axis=1)
test_df = test_df.drop(["Name", "Sex", "Ticket" , "Cabin", "Embarked", "Unknown"], axis=1)
test_df
```

```python
clf.fit(train_df.drop(["Survived"], axis=1), train_df["Survived"])
pre = clf.predict(test_df.drop(["PassengerId"], axis=1))
pre
```

```python
result_df = pd.DataFrame()
result_df["PassengerId"] = test_df["PassengerId"]
result_df["Survived"] = pre
result_df.to_csv("result.csv", encoding="utf-8", index=False)
result_df
```
